```python
blank_img = np.zeros (shape=(512,512,3),dtype=np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img , pt1 = (384,10) , pt2 = (500,150) , color = (255,0,0) , thickness = 20) #thinkness kalinlik oluyor

#pt1 birinci point, eksenlerden 384e 10 noktalarini aliyoruz, color = (255,0,0) ilki red ,green blau sirasiyla gidiyor
#resimde iki nokta belirlemeliyiz,
# bu iki noktayi belirledikten sonra ,kösegen cizilip o kareler tanimlaniyor,

#rectangle(reg-tengııl) dikdörtgen, thickness = -1 yaparsan icini dolduruyor
```

```python
plt.imshow(blank_img)
#yukardaki iki noktayi yani kösegeni yakalayip kareye tamamliyor
#bu yöntemle mesela yüz olan fotoda ,biyigin yerini belirlersen ,oraya koyabilirsin
```

```python
cv2.rectangle(blank_img , pt1 = (200,400) , pt2 = (400,200) , color = (200,200,0) , thickness = 20)
#bu sekilde renleri degistirebiliriz
```

```python
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img , center = (200,400) , radius=20 , color = (200,200,0) , thickness = 8)

# thickness = -1 olunca icini dolduruyor
#radius yari cap veriyor
```

```python
plt.imshow(blank_img)
```

```python
cv2.line(blank_img , pt1 = (0,0) , pt2 = (512,512) , color = (200,200,0) , thickness = 20)

#buda belirledigin iki nokta arasi cizgi ciziyor
```

```python
cv2.line(blank_img , pt1 = (222,25) , pt2 = (400,350) , color = (200,200,0) , thickness = 20)
```

```python
blank_img = np.zeros (shape=(512,512,3),dtype=np.int16)
```

```python
font= cv2.FONT_HERSHEY_SIMPLEX #font ismi
cv2.putText(blank_img, text='Hello',org=(10,500), fontFace=font, fontScale=10, color=(0,255,0),thickness=3, lineType=cv2.LINE_AA)
plt.imshow(blank_img)

# putText yazi kkoy 
# org textin baslangici nerden olsun org =orjin
#resmin baslangicibi belirliyoruz sadece , sigmayinca göstermiyor
```
